### Target Transformation Lookup


| Log Value | Approx Selling Price | Note / Comment |
| :--- | :--- | :--- |
| **11** | ~60,000 | Baseline for low-budget / older hatchbacks. |
| **12** | ~1.6 lakh | Typical price floor for mid-range used sedans. |
| **13** | ~4.4 lakh | Average range for premium hatchbacks or compact SUVs. |
| **14** | ~12 lakh | Entry-level luxury cars or brand-new mid-SUVs. |
| **15** | ~32 lakh | Luxury segment cars (BMW 3 Series, Audi A4, etc.). |
| **16** | ~88 lakh | Supercars, premium imports, or high-end luxury SUVs. |



The original selling price distribution was highly positively skewed due to presence of luxury vehicles and extreme outliers. Applying logarithmic transformation significantly reduced skewness and produced a near-normal distribution, making the data more suitable for regression modeling.

Most values lie between 12.5 and 13.5 that is 3-7 lakh

```python
def brand_segment(brand):
    luxury = ['Lexus', 'Volvo', 'Jaguar', 'BMW', 
              'Land', 'Audi', 'Mercedes-Benz']
    
    mid    = ['MG', 'Jeep', 'Kia', 'Isuzu', 'Force',
              'Toyota', 'Mitsubishi', 'Mahindra', 'Honda',
              'Ford', 'Volkswagen', 'Skoda']
    
    budget = ['Hyundai', 'Renault', 'Nissan', 'Maruti',
              'Datsun', 'Ashok', 'Fiat', 'Tata',
              'Chevrolet', 'Ambassador', 'Daewoo', 'Opel']
    
    if brand in luxury: return 2
    elif brand in mid:  return 1
    else:               return 0

df['brand_segment'] = df['brand'].apply(brand_segment)
df.drop('brand', axis=1, inplace=True)

print(df['brand_segment'].value_counts())
```

```python
from sklearn.model_selection import train_test_split
x=df.drop('log_selling_price',axis=1)
y=df['log_selling_price']
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
print(numerical_cols)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
num_cols=['mileage','engine','max_power','torque','seats','age','log_km_Driven']
x_train[num_cols]=sc.fit_transform(x_train[num_cols])
x_test[num_cols]=sc.transform(x_test[num_cols])
```

```python
#model training
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error,root_mean_squared_error
from sklearn.model_selection import cross_validate
lr=LinearRegression()
cv=cross_validate(lr,x_train,y_train,cv=5,return_train_score=True,scoring={'r2':'r2',
        'mae':'neg_mean_absolute_error',
        'rmse':'neg_root_mean_squared_error'})
print("Train R2 Mean:",
      cv['train_r2'].mean())

print("Validation R2 Mean:",
      cv['test_r2'].mean())

print("Validation R2 Std:",
      cv['test_r2'].std())

print("Validation MAE:",
      -cv['test_mae'].mean())

print("Validation RMSE:",
      -cv['test_rmse'].mean())
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
y_test_actual = np.expm1(y_test)
y_pred_actual = np.expm1(y_pred)
```

```python
from sklearn.metrics import mean_absolute_error
print("MAE:",mean_absolute_error(y_test_actual, y_pred_actual))
print("MSE:",mean_squared_error(y_test_actual, y_pred_actual))
print("R2 Score:",r2_score(y_test, y_pred))
```

```python
residual=y_test_actual-y_pred_actual
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x=y_pred_actual,
    y=residual
)

plt.axhline(y=0, color='red', linestyle='--')

plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residual Plot")

plt.show()
```

Residual analysis indicates that the model errors are
approximately centered around zero, suggesting no major
systematic bias.

However, the residual spread increases for higher predicted
prices, indicating heteroscedasticity. This suggests that
Linear Regression struggles to model expensive vehicles
consistently.

The absence of strong nonlinear residual patterns suggests
that logarithmic transformation improved linearity in the data.

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=100,max_depth=10)
cv_rf=cross_validate(rf,x_train,y_train,cv=5,scoring={
        'r2':'r2',
        'mae':'neg_mean_absolute_error',
        'rmse':'neg_root_mean_squared_error'
    },
    return_train_score=True,
    n_jobs=-1
)
print("Train R2 Mean:",
      cv_rf['train_r2'].mean())
print("Validation R2 Mean:",
      cv_rf['test_r2'].mean())
print("Validation R2 Std:",
      cv_rf['test_r2'].std())
print("Validation MAE:",
      -cv_rf['test_mae'].mean())
print("Validation RMSE:",
      -cv_rf['test_rmse'].mean())
```

## insight 

Among all the implemented regression models, Random Forest Regressor achieved the best overall performance with the highest R² score of approximately 0.91. The model effectively captured non-linear relationships between vehicle features and selling price while reducing overfitting through ensemble learning. Compared to Linear Regression and Decision Tree Regressor, Random Forest produced more stable and accurate predictions. XGBoost also delivered strong performance with an R² score close to 0.90, indicating excellent predictive capability. However, Random Forest slightly outperformed XGBoost on the current dataset and demonstrated better generalization overall.


Despite strong overall performance, both Random Forest and XGBoost struggled to accurately predict high-end luxury vehicles. This limitation is mainly caused by dataset imbalance, where premium vehicles are significantly underrepresented compared to economy and mid-range cars. The absence of luxury-specific features and extreme price variance also contributed to higher prediction errors for premium vehicle segments.
